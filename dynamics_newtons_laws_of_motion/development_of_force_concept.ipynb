# Development of Force Concept

**Dynamics** is the study of how forces act on objects and systems to cause them move. An intuitive understanding of force comes from the idea of *pull and push*. A pull or push has a magnitude as well as a direction. Our everyday experience also suggests that pull and push can stack up to act on an object. Therefore, we can define a force as a vector quantity. Should other quantitative definitions of different forces be developed later.

## Free-body Diagram

*Free-body diagram* can be used to illustrate how the external forces act on a body, regardless of the surrounding environment and the body. The body is represented as an isolated point (free body), and only the external forces acting on the body are shown. Free-body diagrams are very useful in analyzing forces acting on a system.
